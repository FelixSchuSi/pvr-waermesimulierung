```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import seaborn as sns
sns.load_dataset('titanic')
```

```python
titanic = sns.load_dataset('titanic')
fig, ax = plt.subplots(figsize=(6,4), dpi=200)
ax.hist('age', data=titanic, bins=np.arange(0,85,5),color='#56B4E9', rwidth=0.95, alpha=0.9)
ax.set(xlabel='age (years)', ylabel='count')
ax.set_axisbelow(True)
ax.tick_params(axis='y', length=0)
grid_color= ax.get_xgridlines()[0].get_color()
ax.tick_params(axis='x', color=grid_color)
ax.grid(axis='y')
ax.set_xlim(0, 80)

for spine in ax.spines.values():
    spine.set_visible(False)
```

```python
fig.savefig('test.pdf', dpi=300, bbox_inches='tight', pad_inches=0, format='pdf')
```
